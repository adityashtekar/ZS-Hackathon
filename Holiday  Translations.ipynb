```python
import pandas as pd

holiday_df = pd.read_excel('./dataset/holidays.xlsx')
```

```python
print(holiday_df)
print(holiday_df.info())
print(holiday_df.describe())
```

```python
from googletrans import Translator

# holiday_df['Holiday_Trans'] = holiday_df['Holiday'].apply(lambda x:translator.translate(x,src='da',dest='en'))
holiday_df['Holiday_Trans'] = "Default"
for i in range(0,652):
    if(i%50==0):
        translator = Translator()
    x = holiday_df.loc[i,'Holiday']
    if(translator.detect(x)!='en'):
        holiday_df.loc[i,'Holiday_Trans'] = translator.translate(x,dest='en').text
    else:
        holiday_df.loc[i,'Holiday_Trans'] = x.text
    print(str(i)+ " " + str(holiday_df.loc[i,'Holiday_Trans']))
```

```python
print(holiday_df['Holiday_Trans'].value_counts())
```

```python
holiday_df['Holiday_Trans'] = holiday_df['Holiday_Trans'].str.replace(r"\[.*\]","")
holiday_df['Holiday_Trans'] = holiday_df['Holiday_Trans'].str.replace(r"2. ","")
holiday_df['Holiday_Trans'] = holiday_df['Holiday_Trans'].str.replace(r"\"","")
holiday_df['Holiday_Trans'] = holiday_df['Holiday_Trans'].str.replace(r"}","")
holiday_df['Holiday_Trans'] = holiday_df['Holiday_Trans'].str.replace(r"\'s","")
holiday_df['Holiday_Trans'] = holiday_df['Holiday_Trans'].str.replace(r"\'","")
holiday_df['Holiday_Trans'] = holiday_df['Holiday_Trans'].str.replace(r".","")
```
